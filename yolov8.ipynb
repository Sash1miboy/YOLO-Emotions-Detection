```python
from ultralytics import YOLO
```

```python
# Dataset YAML Path
data_path = r"C:\Users\User\Desktop\New-Preprocessed-Data\data.yaml"

# Project Name
project_name = "emotion_detection_yolov8_comparison"
```

```python
# Training Parameter
train_params = {
    "epochs": 120,
    "patience": 20,
    "batch": 32,
    "imgsz": 640,
    "save": True,
    "save_period": 20,
    "device": 0,
    "pretrained":True,
    "optimizer": "AdamW",
    "seed": 3407,
    "cos_lr": True,
    "close_mosaic": True,
    "freeze": 5,
    "lr0": 0.001,
    "lrf": 0.01,
    "momentum": 0.9,
    "val": True,
    "plots": True,

    # Augmentations
    "hsv_h": 0.015, # Variasi Random Color
    "hsv_s": 0.7,   # Saturation Randomness
    "hsv_v": 0.4,   # Brightness Randomness
    "degrees": 10.0,
    "translate": 0.1,
    "scale": 0.5,
    "fliplr": 0.5,
    "mosaic": 1.0,
}
```

```python
# Validation Parameter
val_params = {
    "imgsz": 640,
    "batch": 32,
    "save_json": True,
    "device": 0,
    "plots": True,
    "split": 'val',
    "verbose": True,
    "save_txt": True,
    "save_conf": True,
}
```

```python
# Testing Parameter
test_params = {
    "imgsz": 640,
    "batch": 32,
    "save_json": True,
    "device": 0,
    "plots": True,
    "split": 'test',
    "verbose": True,
    "save_txt": True,
    "save_conf": True,
}
```

## YOLOv8n (Nano Size)

### Train

```python
model_n = YOLO("yolov8n.pt")
```

```python
n_result = model_n.train(
    data=data_path,
    project=project_name,
    name="yolov8n_train_results",
    **train_params
)
```

### Validation

```python
model_n_val = YOLO("path/to/best.pt")
```

```python
val_n_result = model_n_val.val(
    data=data_path,
    project=project_name,
    name="yolov8n_validation_results",
    **val_params
)
```

### Testing

```python
model_n_test = YOLO(r"E:\thesis-model\YOLO-Emotions-Detection\emotion_detection_yolov8_comparison\yolov8n_train_results\weights\best.pt")
```

```python
test_n_result = model_n_test.val(
    data=data_path,
    project=project_name,
    name="yolov8n_testing_results",
    **test_params
)
```
